```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files18 = ['2018/JC-201803-citibike-tripdata.csv', '2018/JC-201804-citibike-tripdata.csv', '2018/JC-201805-citibike-tripdata.csv',
              '2018/JC-201806-citibike-tripdata.csv', '2018/JC-201807-citibike-tripdata.csv', '2018/JC-201808-citibike-tripdata.csv',
              '2018/JC-201809-citibike-tripdata.csv', '2018/JC-201810-citibike-tripdata.csv', '2018/JC-201811-citibike-tripdata.csv',
              '2018/JC-201812-citibike-tripdata.csv', '2018/JC-201801-citibike-tripdata.csv', '2018/JC-201802-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df18_list = []

# Loop through each file in the list
for file in files18:
    # Read the file into a dataframe
    df18 = pd.read_csv(file)

    # Append the dataframe to the list
    df18_list.append(df18)

# Concatenate all dataframes in the list into one dataframe
df18 = pd.concat(df18_list)

output_file_path = 'tableau_data/combined_data_18.csv'
df18.to_csv(output_file_path, index=False)

# Show dataframe
df18.head()
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df18, df20, df22], axis=0)

# Display the results
combined_df.head()
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
combined_df['gender'] = combined_df['gender'].map(mapping)

# Show DataFrame
combined_df
```
